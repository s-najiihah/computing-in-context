Importing and loading our data 

```python
import pandas as pd

df = pd.read_csv("ridership_ktmb_daily.csv")
df.head()
```

Checking how many rows we have

```python
num_rows = df.shape[0]
print("Number of rows:", num_rows)
```

Computing mean, median and mode of ridership for 2025

```python
#Assigning the numeric column to a variable

ridership = df["ridership"]


# Compute statistics
mean_val = ridership.mean()
median_val = ridership.median()
mode_val = ridership.mode()[0]  # mode() may return multiple values

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

Without using pandas

```python
#Reading CSV and extracting ridership data
ridership = []
with open("ridership_ktmb_daily.csv", "r") as f:
    next(f)  # skip header
    for line in f:
        parts = line.strip().split(",")
        if parts[2]:  # skip empty values
            ridership.append(int(parts[2]))
```

```python
#Computing mean

total = sum(ridership)
count = len(ridership)
mean_val = total / count
print("Mean:", mean_val)
```

```python
#Computing median
sorted_r = sorted(ridership)
n = len(sorted_r)
if n % 2 == 1:  
    median_val = sorted_r[n // 2]
else: 
    median_val = (sorted_r[n // 2 - 1] + sorted_r[n // 2]) / 2
print("Median:", median_val)
```

```python
#Computing mode
counts = {}
for val in ridership:
    counts[val] = counts.get(val, 0) + 1
max_count = max(counts.values())
modes = [key for key, val in counts.items() if val == max_count]
mode_val = modes[0]  # pick first mode if multiple
print("Mode:", mode_val)
```
